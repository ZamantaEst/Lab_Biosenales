# **Promedio por canales**

Función de energía promedio

```python
def energia_promedio(signal):
  canales, puntos, epocas= signal.shape
  energia_epoca=np.sum(signal**2,axis=1)
  energias_prom=np.mean(energia_epoca, axis=1)
  return energias_prom
```

Calcule la energía de cada canal promediada por épocas para cada sujeto, esto para ambos grupos poblacionales. Guarde esta información en un DataFrame de columnas ‘canal’ y filas ‘#sujeto’ con los valores de energía calculados, un DataFrame para cada grupo poblacional.

```python
#Energia promedio control
energia_prom_control=[]
for i in range(0,len(archivos_c)):
    #ruta de un archivo
    sujeto_c=(directorio_actual + "/control" + "/" + archivos_c[i]);
    #1 CARGAR EL ARCHIVO
    data_control = sio.loadmat(sujeto_c);
    data_control = data_control['data']
    energia_prom_control.append(energia_promedio(data_control))


#Energia promedio parkinson
energia_prom_parkinson=[]
for i in range(0,len(archivos_p)):
    #ruta de un archivo
    sujeto_p=(directorio_actual + "/parkinson" + "/" + archivos_p[i]);
    #1 CARGAR EL ARCHIVO
    data_parkinson = sio.loadmat(sujeto_p);
    data_parkinson= data_parkinson['data']
    energia_prom_parkinson.append(energia_promedio(data_parkinson))

```

#**Pruebas estadisticas**

Determine si existe diferencia estadística entre canales de cada grupo de sujetos a través de una prueba t. Compruebe los supuestos necesarios para realizar una prueba t, esto es: Normalidad de la variable, independencia (se asume que los grupos son independientes), y homocedasticidad (use una prueba de
Levene), finalmente realice la prueba t para determinar si existen diferencias entre los canales entre grupos de sujetos. De no cumplirse los requisitos, realice entonces un análisis no paramétrico (prueba U de Mann-Whitney). Este numeral tiene como objetivo identificar los canales que entregan información diferencial entre pacientes Sanos y con enfermedad de Parkinson.

#Prueba de normalidad
H0= los datos provienen de una distribución normal

H1= los datos no provienen de una distribución normal

Si valor p<0.05 rechazo H0

```python
print('Normalidad poblacion de control')
for i in range(0,8):
  print('Valor p canal '+str(i+1)+': '+str((sis.shapiro(dataFrame_c['Canal '+str(i+1)]))[1]))

print('\nNormalidad poblacion de parkinson')
for i in range(0,8):
  print('Valor p canal '+str(i+1)+': '+str((sis.shapiro(dataFrame_p['Canal '+str(i+1)]))[1]))
```

Como el valor p de los canales de cada poblacion son menores a 0.5, se rechaza H0, por tanto los datos no provienen de una distibucion normal

#Prueba de homocedasticidad

H0= hay homocedasticidad entre los dos datos

H1= hay heterocedasticidad entre los dos datos

Si valor p<0.05 rechazo H0

```python
#Homocedasticidad comparando cada canal de las dos poblaciones
for i in range (0,8):
  print('Valor p canal '+str(i+1)+ ': ' +str((sis.levene(dataFrame_c['Canal '+str(i+1)],dataFrame_p['Canal '+str(i+1)]))[1]))
```

El valor p entre cada canal de ambos grupos es mayor a 0.05, se acepta H0. Hay homocedasticidad entre los dos datos (los datos no presentan dferencia significativa en sus varianzas). Sin embargo, es evidente que en algunos canales el valor p es mucho menor en comparación a otros, como lo son el canal 4, 6, 7 y 8 lo que indica que a pesar de no tener diferencias significativas en sus varianzas, esta diferencia si es mayor comparada con los canales 1, 2, 3 y 5.

#Prueba de independencia

H0= Las grupos son independientes

H1= Las grupos no son independientes

```python
#Independencia comparando cada canal de las dos poblaciones
for i in range (0,8):
  print('Valor p canal '+str(i+1)+ ': ' +str((sis.chi2_contingency(dataFrame_c['Canal '+str(i+1)],dataFrame_p['Canal '+str(i+1)]))[1]))
```

Como el valor p de cada canal es mayor a 0.05, se acepta H0 y se afirma que los grupos son independientes

#Prueba U de Mann-Whitney

Como no se cumplieron los supuestos de homocedasticidad y de normalidad, no se puede hacer la prueba t y se procede con la U de Mann-Whitney.

H0=Las medianas de las muestras no tienen diferencias significativas.

H1= Las medianas de las muestras tienen diferencias significativas.

```python
#U de Mann-Whitney comparando cada canal de las dos poblaciones
for i in range (0,8):
  print('Valor p canal '+str(i+1)+ ': ' +str((sis.mannwhitneyu(dataFrame_c['Canal '+str(i+1)],dataFrame_p['Canal '+str(i+1)]))[1]))
```

Como el valor p para ambas poblaciones fue mayor de 0.05, se acepta H0 y por lo tanto, las medianas de los datos no tienen diferencias significativas. A pesar de esto, se observa que los valores p obtenidos en los canales 4, 6, 7 y 8 son menores a los obtenidos en los canales 1, 2, 3 y 5, lo que puede indicar que la diferencia en las medianas de los canales 4, 6, 7 y 8 de los dos grupos presenta una mayor diferencia en comparación a los otros canales.
