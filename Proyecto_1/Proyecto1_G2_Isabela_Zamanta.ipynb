1. Consultar para enfermedad de Parkinson, Huntington y Esclerosis Lateral Amiotrófica cómo se diagnostican y el reto que existe para el diagnóstico temprano de dichas enfermedades (5%).  

**Parkinson**

La enfermedad de Parkinson es de diagnóstico clínico. Un neurólogo realiza un proceso de anamnesis (exploración clínica) detallada, la revisión de la historia clínica, exploración física y revisión de la sintomatología del paciente para realizar el diagnostico. Actualmente, no hay ningun examen con el que se pueda realizar el diagnóstico, pero, si se pueden realizar algunos estudios como tomografía computarizada o resonancia magnética nuclear para descartar enfermedades que tengan similar sintomatología.  

Otro estudio, es el DaTscan (imagen de transportadores de dopamina), este permite la visualización del sistema dopaminérgico en el cerebro y si tiene un resultado positivo, puede ayudar a confirmar el diagnóstico de la enfermedad de Parkinson, pero no distingue entre esta enfermedad y otros parkinsonismos atípicos. [1]

En algunos casos se administran tratamientos para la enfermedad de Parkinson para ver su utilidad y así corroborar el diagnóstico. [2]

Al inicio de la enfermedad, los síntomas son poco perceptibles lo que dificulta su detección temprana, además, su diagnóstico es complejo es personas mayores dado que el deterioro causado por el envejecimiento puede provocar síntomas similares a los del Parkinson. [2]

**Huntington**

El diagnóstico de la enfermedad de Huntington se sospecha por la presencia de sintomatología característica, además de antecedentes familiares. Exámenes imagenológicos como la tomografía computarizada y la resonancia magnética cerebral pueden ayudar a determinar el diagnóstico.

Se confirma la enfermedad de Huntington con una prueba genética (dado que la enfermedad se debe a un defecto genético en el gen HTT) [3]

Su detección temprana se dificulta debido a que los sintomas pueden pasar desapercibidos o confundidos con otras afecciones. También se dificulta porque los cambios estructurales característicos de la enfermedad y que pueden observan por exámenes de imagen no son evidentes en las primeras etapas de la enfermedad. [4]

**Esclerosis Lateral Amiotrófica**

No existe ninguna prueba específica que permita su detección. El diagnóstico de la esclerosis lateral amiotrófica (ELA) se realiza en conjunto con pruebas clínicas (evaluación de los síntomas y de la historia clínica del paciente), pruebas neurológicas y de imagen como la electromiografía (EMG), resonancia magnética (MRI), estudios de líquido cefalorraquídeo (LCR), estudios de conducción nerviosa y todos estos sirven para descartar otras enfermedades. [5][6]

Su diagnóstico en estados tempranos es difícil, comenzando porque la sintomatología es variable en cada persona y puede confundirse con síntomas de otras enfermedades, además, puede tener afectaciones que inicialmente son sutiles y que pueden pasar desapercibidas como espasmos musculares, calambres, rigidez o debilidad. [7]

La entropía de permutación es una medida que cuantifica la complejidad de una serie temporal basada en el orden de los valores de la señal. Matemáticamente, se define a partir de la entropía de Shannon, calculada utilizando la siguiente fórmula[12]:

La entropía de permutación evalúa la relación de orden entre los valores de una señal, cuantificando la probabilidad de que un estado siga a otro. Para su cálculo, se requieren varios parámetros de entrada, como la longitud de la serie temporal (N), el radio de tolerancia (r) y la longitud del patrón (m), también conocida como ventana del patrón. Estos parámetros son críticos para la precisión del cálculo, por lo que deben seleccionarse con cuidado. Por ejemplo, el radio de tolerancia determina si los patrones en la serie se consideran similares: un valor de r demasiado pequeño limitaría las coincidencias, mientras que uno excesivamente grande podría agrupar patrones dispares, distorsionando los resultados. Asimismo, un valor bajo de m podría no captar la complejidad de la señal, mientras que un valor alto podría dificultar la identificación de permutaciones únicas.

El valor de la entropía de permutación proporciona información relevante sobre la predictibilidad de la serie. Un valor cercano a 0 indica que la serie es altamente predecible, mientras que un valor alto sugiere mayor complejidad y desorden, lo que implica la aparición de nueva información a medida que la serie avanza.

La entropía de permutación se basa únicamente en el orden de los valores, sin considerar las diferencias numéricas absolutas entre ellos. Esto significa que dos secuencias que presenten el mismo patrón de aumento o disminución se consideran similares en términos de permutación, aunque sus valores absolutos sean muy diferentes. Esta característica hace que la entropía de permutación sea útil en contextos donde el orden de los eventos es más importante que sus magnitudes.

En bioingeniería, la entropía de permutación se utiliza para el análisis de señales fisiológicas, como el electroencefalograma (EEG) o el electrocardiograma (ECG). En el caso del EEG, puede emplearse para evaluar los niveles de actividad cerebral en distintos estados, como el sueño o la anestesia. Un EEG con baja entropía podría señalar un estado de rigidez cerebral, observado en patologías neurológicas como la epilepsia. En el ECG, la entropía de permutación puede analizar la variabilidad de la frecuencia cardíaca, proporcionando información sobre el funcionamiento del sistema cardiovascular.[12][13]

4. Crear una función que reciba una señal de marcha y calcule el coeficiente de variación (CV) y la **SD tal como se define en el artículo de referencia**

```python
def Rutina2 (senal):
  SD=np.std(senal)
  CV=(100*SD)/abs(np.mean(senal))
  return CV,SD
```

5. Crear una función que reciba una señal de marcha y permita calcular la entropía de permutación (10%)

```python
def shannon_entropy(x):
  p , _ = np.histogram(x,bins='fd') #'fd' (estimador de Freedman Diaconis)para el tamaño de los datos en el histograma
  p = p/np.sum(p) # normalizacion de los datos se divide entre la cantidad de todos los datos de p
  h = -p*np.log2(p) #shannon
  h[np.isnan(h)] = 0.0 #elimina los valores not a number errores matematicos como log(0)
  H = np.sum(h) #entropia total
  return H

#shannon_entropy(a[:,1])
```

# Prueba de homocedasticidad

H0= hay homocedasticidad entre los dos ciclos

H1= hay heterocedasticidad entre los dos ciclos

```python
for i in range(2,14):
  print(f"Valor p de homocedasticidad {dataFrame_c.iloc[:,i].name}")
  print('Valor p: '+str(round((sis.levene(dataFrame_c.iloc[:,i].values,dataFrame_p.iloc[:,i].values,dataFrame_h.iloc[:,i].values,
                                         dataFrame_ela.iloc[:,i].values))[1],4)))
```

Como el valor p de homocedasticidad es mayor a 0.05 para todos los casos menos para EP intervalo balanceo pie izquierdo, se cumple la H0 y por tanto hay homocedasticidad entre todos los grupos en cada columna, pero para EP intervalo balanceo pie izquierdo hay heterocedasticidad.

#Prueba de independencia

H0= Las grupos son independientes

H1= Las grupos no son independientes



```python
for i in range(2,14):
  print(f"Valor p de homocedasticidad {dataFrame_c.iloc[:,i].name}")
  datosa=np.concatenate((dataFrame_c.iloc[:,i].values,dataFrame_p.iloc[:,i].values))
  print('Valor p: '+str(round((sis.chi2_contingency(datosa,dataFrame_h.iloc[:,i].values,dataFrame_ela.iloc[:,i].values))[1],4)))
```

Como el valor p para la prueba de independencia es mayor a 0.05, se acepta H0 y se concluye que los datos son independientes.

**Conclusión:** Ya que no se cumplieron los supuestos de homocedasticidad y normalidad para todos los datos de todas las poblaciones, no se puede hacer la prueba t y se procede a realizar la prueba ***no paramétrica*** de U Mann-Whitney.

#Prueba U de Mann-Whitney

H0=Las medianas de las muestras no tienen diferencias significativas.

H1= Las medianas de las muestras tienen diferencias significativas.

```python
for i in range(2,14):
  print(f"Valor p de la Prueba U de Mann-Whitney {dataFrame_c.iloc[:,i].name}")
  datosb=np.concatenate((dataFrame_c.iloc[:,i].values,dataFrame_p.iloc[:,i].values))
  datosc=np.concatenate((dataFrame_h.iloc[:,i].values,dataFrame_ela.iloc[:,i].values))
  print('Valor p: '+str(round((sis.mannwhitneyu(datosb,datosc))[1],4)))
```

Como los valores p de cada señal de la prueba de U Mann-Whitney son variados y algunos de los datos aceptan H0 y otros la rechazar.
