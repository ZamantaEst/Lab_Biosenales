Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.


```python
def alumnos(notas):
  lista_n=[]
  for i in notas:
    lista_n.append(float(notas[i]))
    n_min=min(lista_n)
    n_max=max(lista_n)
    n_media=np.mean(lista_n)
    desviacion= np.std(lista_n)

  serie=[n_min,n_max,n_media, desviacion]
  return serie
```

```python
#Uso
dic={ "Pedro": "1.2", "Pepito": "5", "Andrea": "3.2","Sara": "4.2", "Luisa": "2", "Carlos": "2.7","Catalina": "2.3", "Mariana": "4.5"}
alumnos(dic)
```
