#Impulso unitario


```python
def impseq(n0, n1, n2):
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) == 0
 plt.stem(n,x)
 return [x,n]
impseq(0,-2,3)
```

4. Genere la siguiente secuencia
ùë•[ùëõ] = {
ùëü[ùëõ] 0 ‚â§ ùëõ ‚â§ 5
ùëü[ùëõ ‚àí 5] 6 ‚â§ ùëõ ‚â§ 11
ùëü[ùëõ ‚àí 10] 12 ‚â§ ùëõ ‚â§ 17
* Derive la secuencia
* Realice un subplot con las dos secuencias, la original, y la derivada. Describa lo que observa
y el porqu√© no es exactamente la derivada del caso continuo.

```python
n4=np.arange(0,18)
def frampa(n0, n1, n2):
    r,n4=escseq(n0, n1, n2)
    for i in n4:
      if (0<=i and i<=5):
        x = (n4)*r
      elif (6<=i and i<=11):
        x = (n4-5)*r
      elif (12<=i and i<=17):
        x = (n4-10)*r
    return x.astype(int)
```

```python
#Secuencia original
X1=(frampa(0,0,5))
X2=(frampa(5,6,11))
X3=(frampa(10,12,17))
X=np.concatenate((X1, X2, X3))
print(X)
plt.subplot(2,1,1)
plt.subplots_adjust(hspace=0.5) #Crea un espacio entre las gr√°ficas
plt.stem(n4,X,label='Original')
plt.title('Original')
#Secuencia derivada
plt.subplot(2,1,2)
derivative = np.gradient(X)
plt.stem(n4, derivative)
plt.title('Derivada')
```
