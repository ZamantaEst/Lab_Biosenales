#Impulso unitario


```python
def impseq(n0, n1, n2):
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) == 0
 plt.stem(n,x)
 return [x,n]
impseq(0,-2,3)
```

#Escalon unitario
1. Implemente una función en Python que permita generar una función escalón unitario, definida en un intervalo 𝑛1≤𝑛0≤𝑛


```python
def escseq(n0, n1, n2):
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) >= 0
 return [x,n]
 return x.astype(int)

```

```python
y=escseq(0,-2,3)
print(y[0])
plt.stem(n,y[0])
```

4. Genere la siguiente secuencia
𝑥[𝑛] = {
𝑟[𝑛] 0 ≤ 𝑛 ≤ 5
𝑟[𝑛 − 5] 6 ≤ 𝑛 ≤ 11
𝑟[𝑛 − 10] 12 ≤ 𝑛 ≤ 17
* Derive la secuencia
* Realice un subplot con las dos secuencias, la original, y la derivada. Describa lo que observa
y el porqué no es exactamente la derivada del caso continuo.

```python
n4=np.arange(0,18)
def frampa(n0, n1, n2):
    r,n4=escseq(n0, n1, n2)
    for i in n4:
      if (0<=i and i<=5):
        x = (n4)*r
      elif (6<=i and i<=11):
        x = (n4-5)*r
      elif (12<=i and i<=17):
        x = (n4-10)*r
    return x.astype(int)
```

```python
#Secuencia original
X1=(frampa(0,0,5))
X2=(frampa(5,6,11))
X3=(frampa(10,12,17))
X=np.concatenate((X1, X2, X3))
print(X)
plt.subplot(2,1,1)
plt.subplots_adjust(hspace=0.5) #Crea un espacio entre las gráficas
plt.stem(n4,X,label='Original')
plt.title('Original')
#Secuencia derivada
plt.subplot(2,1,2)
derivative = np.gradient(X)
plt.stem(n4, derivative)
plt.title('Derivada')
```
