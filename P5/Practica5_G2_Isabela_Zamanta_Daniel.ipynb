Las líneas anteriores si aplican un filtro. Primero se calcula la transformada rápida de Fourier de la señal y se guarda en F1 que representa la señal en el dominio frecuencial, luego se crea un arreglo F2 de ceros de igual longitud que la señal con la TFF (F1). Luego en el arreglo F2 se copian unas frecuencias específicas (fecuencias de interés) de la señal F1 y esto hace que se supriman o se omitan las demás frecuencias. Finalmente se realiza la TFF inversa para volver al dominio del tiempo. Esto es un filtro debido a que modifica las frecuencias contenidas en la señal de forma selectiva y esta modificación en el dominio de la frecuencia cambia características de la señal en el dominio del tiempo.

Para escoger el ancho de ventana (hanning) utilizada en Welch  se tiene en cuenta la longitud de los datos, pues esta será el límite de ancho de banda. Se considera que un valor "bajo" nos limita en cuanto a la resolución obtenida en el espectro encontrado, mientras que un valor muy alto genera mas fluctuacion en las estimaciones .

El solapamiento utilizado se determina en funcion del porcentaje del ancho de ventana que se va a superponer, por lo cuál, aplicar un valor superior al 50% causaría un sobre promediado de muestras, perdiendo parte de la información al reducir demasiado los y aumentar el valor de los valles de la señal en igual medida. Por lo cual, se considera este un buen valor para aprovechar el funcionamiento de la rutina welch para mejorar la resulución y aumentar el número de promedios utilizados.
